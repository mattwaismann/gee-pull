```python
import ee 
from datetime import datetime
from datetime import timedelta
import os
import geemap
import geopandas as gpd
import ee_helpers
```

```python
# Trigger the authentication flow
ee.Authenticate()

# Initialize the library
ee.Initialize()
```

```python
read_dir = os.path.join(os.path.expanduser('~'),"Desktop","projects","glims-processor","output")
read_file = "glims_3000_bb.shp"
df = gpd.read_file(os.path.join(read_dir,read_file))
df = df.iloc[2:5,:]
```

```python
out_dir = os.path.join(os.path.expanduser('~'),"Desktop","projects","gee-pull", "data","training")

for row in df.iterrows():
    glims_id = row[1].glac_id
    bounding_box = eval(row[1].bboxes)
    date = row[1].src_date.split("T")[0]
    date = datetime.strptime(date, '%Y-%m-%d')

    start_date = date - timedelta(days = 1000)
    end_date = date + timedelta(days = 1000)

    ee_helpers.export_landsat_five_images(glims_id,
                                          bounding_box,
                                          start_date,
                                          end_date,
                                          out_dir,
                                          cloudy_pixel_percentage = 80)
```
